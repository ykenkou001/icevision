```python
import numpy as np
import cv2
import pathlib
import glob

################################################
import columndetector
import girderbeamdetector
from columndetector.column_detector import ColumnDetector
from girderbeamdetector.girder_beam_detector import GirderBeamDetector
################################################


column_input_dir = '../サンプルデータ/柱/'
girder_beam_input_dir = '../サンプルデータ/梁/'
column_output_dir = '../可視化画像/柱/'
girder_beam_output_dir = '../可視化画像/梁/'

################################################
column_weight_file_path = 'weights/column/checkpoint.pt'
girder_beam_weight_file_path = 'weights/girder_beam/checkpoint.pt'
################################################
```

```python
def draw_rects(img, rects, color):
    for rect in rects:
        rect = list(map(int, rect))
        x1, y1, x2, y2 = rect
        img = cv2.rectangle(img, (x1, y1), (x2, y2), color, 2)
    return img
```

```python
# 柱の取得
column_path_list = glob.glob(column_input_dir + '*.png')

################################################
column_detector_instance = ColumnDetector(column_weight_file_path)
################################################

for column_path in column_path_list:
  img = cv2.imread(column_path)
  file_name = pathlib.Path(column_path).name
  save_path = column_output_dir + file_name
  
  ################################################
  column_boxes = column_detector_instance.main(img)
  ################################################

  img = draw_rects(img, column_boxes, (255, 0, 0))
  cv2.imwrite(save_path, img)

```

```python
# 梁の取得
girder_beam_path_list = glob.glob(girder_beam_input_dir + '*.png')

################################################
girder_beam_detector_instance = GirderBeamDetector(girder_beam_weight_file_path)
################################################

for girder_beam_path in girder_beam_path_list:
  img = cv2.imread(girder_beam_path)
  file_name = pathlib.Path(girder_beam_path).name
  save_path = girder_beam_output_dir + file_name
  
  ################################################
  girder_beam_boxes = girder_beam_detector_instance.main(img)
  ################################################

  img = draw_rects(img, girder_beam_boxes, (255, 0, 0))
  cv2.imwrite(save_path, img)
```
